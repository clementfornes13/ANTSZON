Fonctions utile pour l'algo

```python
def TakeWeight(elem):
    #On prend le poids du noeux
    return elem[2]

def SortGraph(Graph):
    #On trie le graphe par ordre croissant des poids
    #Graph est donc une liste de tuples (A,B,X)
    NewGraph = []
    for element in Graph :
        if element[0] >= element[1] :
            NewGraph.append((element[1],element[0],element[2]))
        else :
            NewGraph.append(element)
    NewGraph.sort(key=TakeWeight)
    #print("SortGraph/NewGraph:",NewGraph)
    return NewGraph

def BiggestNode(graph):
    #On renvoie le plus grand noeux du graphe
    return max([element[0] for element in graph]+[element[1] for element in graph])

def HaveCycle(graph):
    #print("$$$$$$$")
    #print("HaveCycle/Graph:",graph)
    #On suppose que le graphe est connexe et tiée par ordre croissant des poids
    Profondeur = dict()

    def dfs(node,depth=0):
        #On prend les voisins du noeux
        Voisins = Voisin(graph,node)
        #print("dfs/Voisins:",Voisins)
        #print("dfs/Node:",node)
        #print("dfs/Depth:",depth)
        #print("dfs/Profondeur:",Profondeur)
        #On vérifie que tous les noeux ont été visités
        #On chercher un cycle
        for Lien in Voisins:
            if Lien in Profondeur :
                if Profondeur[Lien] != depth - 1 :
                    return True
            else :
                node = Lien[0] if Lien[0] != node else Lien[1]
                Profondeur[Lien] = depth
                dfs(node, depth + 1)
        return False
            


    firstnode = graph[0][0]
    Resultat = dfs(firstnode)
    return Resultat


def Connexe(Graph):
    #On vérifie que le graphe est connexe
    #On considère que les noeux varient de 0 à n-1
    #
    noeux = {element[0] for element in Graph}
    if len(noeux) == max(noeux)+1 :
        return True
    return False

def TailleGraph(Graph) :
    #print(Graph)
    return sum([element[2] for element in Graph])

def Nb_Nodes(Graph) :
    #On renvoie le nombre de noeux
    Nodes = set()
    for element in Graph :
        Nodes.add(element[0])
        Nodes.add(element[1])
    return len(Nodes)

def CheckFullGraph(Graph,Sous_Graph) :
    #On vérifie que le graphe passe par toutes les sommets
    Nb_sommets_graph = Nb_Nodes(Graph)
    Nb_sommets_sous_graph = Nb_Nodes(Sous_Graph)
    if Nb_sommets_graph == Nb_sommets_sous_graph :
        return True
    return False


def Voisin(Graph,Noeux) : #Trustable
    #On renvoie la liste des voisins d'un noeux
    Voisins = []
    for element in Graph :
        if element[0] == Noeux or element[1] == Noeux :
            Voisins.append(element)
    #print("Voisin/Graph:",Graph)
    #print("Voisin/Noeux:",Noeux)
    #print("Voisin/Voisins:",Voisins)
    return Voisins

def Algo_Kruskal(Graph) :
    #On applique l'algorithme de Kruskal
    #On renvoie le graphe des arêtes de poids minimum
    Graph = SortGraph(Graph)
    if not Connexe(Graph):
        return "Graphe non connexe"
    #On prend les arêtes de poids minimum
    Arbre = []
    for element in Graph :
        #print("/////////")
        #print("Algo_Kruskal/Arbre:",Arbre)
        Resultat = HaveCycle(Arbre+[element])
        #print("HaveCycle(Arbre+[element]) :",test)
        #print("Algo_Kruskal/element:",element)
        #print("Algo_Kruskal/Resultat:",Resultat)
        #Check = CheckFullGraph(Graph,Arbre+[element])
        #if Check :
            #print("Check:",Check)
        if Resultat == False :
            #print("Algo_Kruskal/element/Resultat:",element)
            Arbre.append(element)
        if Nb_Nodes(Arbre) == Nb_Nodes(Graph) :
            break
    return Arbre


```
